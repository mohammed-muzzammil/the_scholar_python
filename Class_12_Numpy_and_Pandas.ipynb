```python
# mod, remainder, mean, median, standard devian, variance 


a = np.array([10, 20, 30])
b = np.array([3, 5, 7])

print(np.mod(a, b))
print(np.remainder(a,b))

print(np.mean(a))
print(np.median(a))
print(np.std(a))

print(np.var(a))
```

```python
# What is a data frame

# Dataframe is 2-dimesional data structure with rows and columns
# You can think of it like a spreadsheet
```

```python
stu_details = {
    "stu_id": [1,2,3,4,5],
    "stu_name": ["A", "B", "C", "D", "E"],
    "region": ["N", "S", "E", "N", "W"]
}
```

```python
stu_details
```

```python
stu_details = pd.DataFrame(stu_details)
```

```python
stu_details
```

```python
stu_marks = {
    "st_id": [1,2,3,4,6],
    "sub": ["py", "stats", "maths", "sql", "ds"],
    "marks": [12, 12, 13, 14, 15]
}
```

```python
stu_marks = pd.DataFrame(stu_marks)
```

```python
stu_marks
```

```python
pd.merge(stu_details, stu_marks, left_on ='stu_id', right_on='st_id', how='inner')
```

```python
pd.merge(stu_details, stu_marks, left_on ='stu_id', right_on='st_id', how='left')
```

```python
pd.merge(stu_details, stu_marks, left_on ='stu_id', right_on='st_id', how='right')
```

```python
pd.merge(stu_details, stu_marks, left_on ='stu_id', right_on='st_id', how='outer')
```
